# RQ3
####  Which teams have the youngest coaches? Rank all the teams by the age of their coach and show the 10 teams with the youngest coaches. Remember that during a season a team could have more coaches, in that case pick the younger of them. Additionally, show the distirbutions of the ages of all coaches in Premier League, using a boxplot. (Hint: There's an attribute birthDate).

```python
#First of all we import the libraries that we need
import pandas as pd
from datetime import date #for changing the date of the brith to correct way 
```

```python
#read the datasets. we only need coaches and teams
coaches = pd.read_json('coaches.json', orient='columns')
teams = pd.read_json('teams.json', orient='columns')
```

```python
#correcting the date clumn
coaches['birthDate']=pd.to_datetime(coaches['birthDate'])
```

```python
#build a new dataframe with the columns which we need
columns=['firstName','lastName','middleName','shortName','currentTeamId','birthDate','wyId']
coachesData = pd.DataFrame(data=coaches,columns=columns)
coachesData.head(5)
```

```python
#function fo calculate the age
def calculateAge(birthDate): 
    today = date.today() 
    age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day))
    return age
```

```python
#calculate the Age from birthDate and put in a new column (Age) on coachesData dataframe
for i in range (len(coachesData)):
    coachesData.loc[i,'Age']=calculateAge(coachesData.iloc[i]['birthDate'])
del coachesData['birthDate']
coachesData.head(5)
```

```python
#export the name of the country from area and put into new column with name country 
area=teams['area']
data_dict = area.to_dict()
final = (pd.DataFrame(data_dict)).T
teams['country']= final['name']
teams.head(5)
```

```python
#build a new dataframe with the columns which we need
columns=['officialName','country','wyId']
teamsData = pd.DataFrame(data=teams,columns=columns)
```

```python
#rename the column wyId in teamsData beacuse of joining
teamsData=teamsData.rename(columns={"wyId": "currentTeamId"})
teamsData.head(5)
```

```python
#merge teamsData and coachesData
finalData=pd.merge(teamsData, coachesData, on='currentTeamId')
finalData.head(5)
```

```python
#showing only England country in our finalData
finalData = finalData[finalData.country=='England']
```

```python
#sort the finalData by Age of the coaches
finalData = finalData.sort_values('Age')
```
