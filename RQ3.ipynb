```python
import pandas as pd
from datetime import date
```

```python
coaches = pd.read_json('coaches.json', orient='columns')
teams = pd.read_json('teams.json', orient='columns')
```

```python
coaches['birthDate']=pd.to_datetime(coaches['birthDate'])
```

```python
coaches.sort_values(by='birthDate', ascending=False)
```

```python
columns=['firstName','lastName','middleName','shortName','currentTeamId','birthDate','wyId']
```

```python
coachesData = pd.DataFrame(data=coaches,columns=columns)
```

```python
coachesData.head(5)
```

```python
def calculateAge(birthDate): 
    today = date.today() 
    age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day))
    return age
```

```python
for i in range (len(coachesData)):
    coachesData.loc[i,'Age']=calculateAge(coachesData.iloc[i]['birthDate'])
```

```python
#del coachesData['birthDate']
#del coachesData.iloc[0]
coachesData.head(5)
```

```python
columns=['officialName','wyId']
```

```python
teamsData = pd.DataFrame(data=teams,columns=columns)
```

```python
teamsData=teamsData.rename(columns={"wyId": "currentTeamId"})
teamsData.head(5)
```

```python
finalData=pd.merge(teamsData, coachesData, on='currentTeamId')
finalData.head(5)
```

```python
finalData = finalData.sort_values('Age')
```
